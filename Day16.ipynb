```python
plt.subplot(2,2,1)
sns.histplot(df["age"])
plt.subplot(2,2,2)
sns.histplot(df["bmi"])
plt.subplot(2,2,3)
sns.histplot(df["children"])
plt.subplot(2,2,4)
sns.histplot(df["expenses"])
plt.tight_layout()
plt.show()
```

```python
sns.histplot(df["age"])
plt.title("Age Distribution")
plt.xlabel("AGE")
plt.ylabel("Numbers")
plt.show()
sns.histplot(df["bmi"])
plt.title("Bmi Distribution")
plt.xlabel("BMI")
plt.ylabel("Numbers")
plt.show()
sns.histplot(df["children"])
plt.title("Children Distribution")
plt.xlabel("Children")
plt.ylabel("Numbers")
plt.show()
sns.histplot(df["expenses"])
plt.title("Expense Distribution")
plt.xlabel("Expense")
plt.ylabel("Numbers")
plt.show()
```

```python
sns.countplot(x="sex",data=df)
plt.title("Sex Count")
plt.xlabel("Sex")
plt.ylabel("Count")
plt.show()
sns.countplot(x="smoker",data=df)
plt.title("Smoker Count")
plt.xlabel("Smoker")
plt.ylabel("Count")
plt.show()
sns.countplot(x="region",data=df)
plt.title("Region Count")
plt.xlabel("Region")
plt.ylabel("Count")
plt.show()
```

```python
corr=df.corr(numeric_only=True)
sns.heatmap(corr,annot=True,cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
print("Average Charges:")
print(df["expenses"].mean())
print("Maximum Charges:")
print(df["expenses"].max())
print("Minimum Charges:")
print(df["expenses"].min())
print("Average Charges by Smoker:")
print(df.groupby("smoker")["expenses"].mean())
```

```python
sns.boxplot(x="smoker",y="expenses",data=df)
plt.title("Expenses by Smoker")
plt.xlabel("Smoker")
plt.ylabel("Expenses")
plt.show()
sns.boxplot(x="sex",y="expenses",data=df)
plt.title("Expenses by Sex")
plt.xlabel("Sex")
plt.ylabel("Expenses")
plt.show()
```

```python
#1.What is the average age and BMI of the customers?
print("1.Average Age of Customers:",df["age"].mean())
print(" Average BMI of Customers:", df["bmi"].mean())
#2.How does smoking affect insurance charges?
print("\n2.Average Insurance Charges by Smoker:")
print(df.groupby("smoker")["expenses"].mean())
#3.Which region has the highest number of customers?
print("\n3.Region with Highest Number of Customers:")
print(df["region"].value_counts().idxmax())
#4.Any interesting observations from the visualizations?
print("\n4.Observations:")
print("-Most customers are middle-aged.")
print("-Smokers have much higher insurance charges than non-smokers.")
print("-The region printed above has the highest number of customers.")
print("-The heatmap and boxplots show that smoking has a strong effect on insurance charges.")
```
