```python
plt.subplot(2,2,1)
plt.title("Age Distribution")
plt.xlabel("AGE")
plt.ylabel("Count of Age")
sns.histplot(df["age"])
plt.subplot(2,2,2)
plt.title("Jersey_Number Distribution")
plt.xlabel("Jersey_Number")
plt.ylabel("Count of Jersey_Number")
sns.histplot(df["jersey_number"])
plt.subplot(2,2,3)
plt.title("Height Distribution")
plt.xlabel("Height")
plt.ylabel("Count of Height")
sns.histplot(df["height_cm"])
plt.subplot(2,2,4)
plt.title("Weight Distribution")
plt.xlabel("Weight")
plt.ylabel("Count of Weight")
sns.histplot(df["weight_kg"])
plt.tight_layout()
plt.show()
```

```python

sns.countplot(x="age",data=df)
plt.title("Age Count")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
sns.countplot(x="jersey_number",data=df)
plt.title("Jersey Count")
plt.xlabel("Jersey")
plt.ylabel("Count")
plt.show()
sns.countplot(x="weight_kg",data=df)
plt.title("Weight Count")
plt.xlabel("Weight")
plt.ylabel("Count")
plt.show()
```

```python
num_data=df[['age','height_cm', 'weight_kg']]
corr = num_data.corr()
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap of Numerical Variables")
plt.show()
```

```python
print("Average Height:")
print(df["height_cm"].mean())
print("Maximum Height:")
print(df["height_cm"].max())
print("Minimum Height:")
print(df["height_cm"].min())
```

```python
sns.boxplot(x="age",y="weight_kg",data=df)
plt.title("Weight of Player")
plt.xlabel("Age")
plt.ylabel("Weight")
plt.show()
sns.boxplot(x="age",y="height_cm",data=df)
plt.title("Height of Player")
plt.xlabel("Age")
plt.ylabel("Height")
plt.show()
```

```python
avg_age=df["age"].mean()
avg_height=df["height_cm"].mean()
most_common_jersey =df["jersey_number"].mode()[0]
jersey_count =df["jersey_number"].value_counts().max()
most_common_age =df["age"].mode()[0]
age_count =df["age"].value_counts().max()
print(f"1.Average Age of Players :{avg_age:.2f} years")
print(f"Average Height of Players :{avg_height:.2f} cm")
print(f"\n2.Most Common Jersey Number :{most_common_jersey}")
print(f"Number of Players Wearing It :{jersey_count}")
print(f"\n3.Most Common Age :{most_common_age}")
print(f"Number of Players :{age_count}")
print("\n4. Interesting Observations:")
print("- Most players are around 26 years old.")
print("- Jersey number 1 is the most frequently assigned jersey.")
print("- The average player height is about 182 cm.")
print("- Most players are between 23 and 29 years of age.")
```
